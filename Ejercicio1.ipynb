Dos personas han quedado citas en un determinado lugar. Ambas tienen la misma probabildad de llegar al lugar de la cita en el intervalo de tiempo [0,30].
Determinese la probabilidad  de que el tiempo que una persona espere a la otra  sea como máximo 10.




Simulación

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Generamos los números aleatorios y relizamos el diseño del experimento.

```python
def t(n):

  lista = []
  count = 0 

  for i in range(n):
    p1= random.uniform(0, 30) 
    p2 = random.uniform(0, 30)
    if p1 < p2 and p2 - p1 <= 10:
        count += 1             
        lista.append(count/(i+1))
  return lista 
```

```python
p = t(10000)
```

Obtenemos la gráfica.

```python
plt.plot(p,color="green")
plt.title('Gráfica')
plt.show()
```

Calculamos la media y la desviación estándar.

```python
print(np.mean(p), np.std(p))
```

Repetimos el modelo n veces.

```python
def modelo(n):
    lista = []
    for i in range(n):
      s = t(10000) 
      lista.append(np.mean(s))
    return lista
```

```python
y=modelo(400)
```

```python
plt.hist(y, color="blue")
plt.show()
```

Nuevamente calculamos la media y la desviación estándar.

```python
print(np.mean(y), np.std(y))
```

Encontramos el intervalo de confianza

```python
print(np.mean(y)-(1.96)*((np.var(y))/np.sqrt(50)),
np.mean(y)+(1.96)*((np.var(y))/np.sqrt(50)))
```

Analitica

Intervalo: [0,30]

Ambas personas tienen la probabilidad de llegar en cualquier momento.


Área total del intervalo:
$$a_t=(30)(30)=900$$

Área del encuentro:
$$a_e=900-2\frac{20^2}{2}=500$$

Probabilidad del enceuntro:
$$P_e=\frac{a_e}{a_t}=\frac{500}{900}=\frac{5}{9}=0.5555$$

$\therefore$ la probabilidad de que el tiempo que una persona espere a la otra como maximo 10 minutos es de 55.5%
