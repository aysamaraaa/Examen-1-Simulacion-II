**Primera política de reemplazo**

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Generamos los números aleatorios.

```python
def ha():
    h=0
    a = random.random()
    if 0.0 <= a < 0.049999:
        h = 20
    elif 0.05 <= a < 0.199999:
        h = 30
    elif 0.20 <= a < 0.349999:
        h = 40
    elif 0.35 <= a < 0.549999:
        h = 50
    elif 0.55 <= a <= 0.749999:
        h = 60
    elif 0.75 <= a < 0.899999:
        h = 70
    elif 0.90 <= a <= 1.00:
        h = 80
    return h
```

```python
def c(n): 
    ch=[]
    j=0
    for i in range(n):
      j=j+ha() 
      ch.append(((i+1)*110)/(j))
    return ch
```

```python
x=c(500)
```

Obtenemos la gráfica:

```python
plt.plot(x,color="green")
plt.title('Gráfica')
plt.show()
```

Calculamos la media

```python
np.mean(x)
```

Repetimos el modelo n veces

```python
def model(n): 
    lista=[]
    for i in range(n):
        s = c(500)
        lista.append(np.mean(s)) 
    return lista
```

```python
Y=model(40) 
```

Calculamos la media  la desviación estandar

```python
print(np.mean(Y), np.std(Y))
```

Graficamos nuevamente.

```python
plt.hist(Y, color="blue")
plt.show()
```

Encontramos el intervalo de confianza.

```python
print(
np.mean(Y)-(1.96)*((np.var(Y))/np.sqrt(40)), 
np.mean(Y)+(1.96)*((np.var(Y))/np.sqrt(40)))
```

```python
def c2(n): 
    ch=[]
    j=0
    for i in range(n):
      j=j+ha()
      ch.append(((i+1)*250)/(j)) 
    return ch
```

```python
def m2(n):
    lista=[]
    for i in range(n):
        s = c2(500)
        lista.append(np.mean(s)) 
    return lista
```

```python
print(np.mean(Y2)-(1.96)*((np.var(Y2))/np.sqrt(40)),
np.mean(Y2)+(1.96)*((np.var(Y2))/np.sqrt(40)))
```
